```
import gpt_2_simple as gpt2
import pandas as pd
```

```
data = pd.read_feather('https://github.com/Scytheface/NARV/raw/main/lyrics/lyrics.ft')
with open('lyrics.txt', 'w') as f:
  f.write('\n<|endoftext|>\n'.join(data['lyrics']).lower() + '\n<|endoftext|>\n')
gpt2.download_gpt2(model_name='335M')
```

```
sess = gpt2.start_tf_sess()

gpt2.finetune(sess, 'lyrics.txt', 
              model_name='355M', 
              steps=1000, 
              restore_from='fresh', 
              print_every=10,
              sample_every=200,
              save_every=500)
```

```
gpt2.generate(sess, temperature=0.666)
```
